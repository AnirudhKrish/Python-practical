```python
#method to find the complete path. Input must be a list of lists, each list having 1st number as head.
def Closed_cycles(w):
    import random
    w1=[]
    for i in range(0,len(w)):
        l=w[i]
        w1.append(l[0])
    r = random.randint(0, len(w)-1)
    cyc=w1[r]
    cycle =[cyc]
    m=1
    a=cyc
    while(m==1):
        f=[]
        for j in range(0,len(w)):
            c=w[j]
            if(c[0]==a):
                f=w[j]
                break
        if(len(f)==1):
            m=0
        else:
            r1=random.randint(1,len(f)-1)
            q=f[r1]
            a=q
            cycle.append(a)
            del f[r1]
    o=1
    while(o==1):
        t=0
        s=0
        for i in range(0,len(cycle)):
            y = cycle[i]
            u=0
            for j in range(0,len(w1)):
                if(y==w1[j]):
                    u=j
                    break
            if(len(w[u])!=1):
                t=y
                s=i
                break
        b=[t]
        n=1
        while(n==1):
            f1=[]
            for j in range(0,len(w)):
                c=w[j]
                if(c[0]==t):
                    f1=w[j]
                    break
            if(len(f1)==1):
                n=0
            else:
                r2= random.randint(1,len(f1)-1)
                q1 = f1[r2]
                t=q1
                b.append(t)
                del f1[r2]
        if(len(b)!=1):
            for i in range(1,len(b)):
                cycle.insert(s+i,b[i])
        x=0
        for i in range(0,len(w)):
            z=w[i]
            if(len(z)==1):
                x=x+1
            else:
                x=x
        if(len(w)==x):
            o==0
            break
    fcycle=[]
    for i in range(len(cycle)):
        p= str(cycle[i])
        fcycle.append(p)
    v="->"
    v=v.join(fcycle)
    return v, len(cycle)
```
